```python
import pymysql, re, ngram
import pandas as pd
import Levenshtein

druglist1100 =list(pd.read_csv('/Users/dpaulsen/Documents/Professional/InsightDataScience/drug_list_medilexicon_1100.txt', header=None, names=['names1100'])['names1100'])
druglist1100 = [x.upper() for x in druglist1100]
for i in range(0, len(druglist1100)):
	druglist1100[i] = re.sub('\s*$', '', druglist1100[i]) # remove all trailing spaces


user_id='root'


conn = pymysql.connect(host='127.0.0.1', port=3306, user=user_id, database='RxFx')
c = conn.cursor()
c.execute("USE RxFx")

names = pd.io.sql.frame_query('''SELECT DISTINCT medicinalproduct FROM drugs ORDER BY medicinalproduct''', conn)
namelist = list(names['medicinalproduct'])
```

```python

names_set = set()
names_dict = {}
for i in range(0, len(namelist)):
	original_value = namelist[i]
	new_value1 = original_value
	if (len(new_value1.split()) == 1):
		new_value1 = re.sub('[()]', '', new_value1) # remove ()'s around solo drugs, e.g. (DIGOXIN)
	new_value1 = re.sub('[,\[\/(].*', '', new_value1) # remove anything after a ",","/","[", or "("
	new_value1 = new_value1.replace('TABLETS', '') # remove the word TABLETS
	new_value1 = new_value1.replace('TABLET', '') # remove the word TABLETS
	new_value1 = new_value1.replace('TABS', '') # remove the word TABLETS
	new_value1 = new_value1.replace('FASTAB', '') # remove the word TABLETS
	new_value1 = new_value1.replace(' ZZ ', '') # remove the word TABLETS
	new_value1 = new_value1.replace(' HCL', '') # remove the word TABLETS
	new_value1 = new_value1.replace('ALLERGY', '') # remove the word TABLETS
	new_value1 = new_value1.replace('UNKNOWN', '') # remove the word UNKNOWN
	new_value1 = new_value1.replace('UNSPECIFIED', '') # remove the word UNSPECIFIED
	new_value1 = re.sub('^\s*|[^0-9a-zA-Z\s-]+|\s*$', '', new_value1) # remove non-word characters
	new_value1 = re.sub('[\^]|[\'\`\*]', '', new_value1) # remove dosages: numbers followed by space followed by "MG"
	new_value1 = re.sub('\d+[\s][M][G]', '', new_value1) # remove dosages: numbers followed by space followed by "MG"
	new_value1 = re.sub('\d+[M][G]', '', new_value1) # remove dosages: numbers followed by "MG"
	new_value1 = re.sub('[\s][\d]+|[\d](?=[a-zA-Z])', '', new_value1) # remove digits preceded by space, and digits situated directly between numbers
	#new_value1 = re.sub('\s{2,120}', ' ', new_value1)
	new_value1 = re.sub('\s*$', '', new_value1) # remove all trailing spaces
	split_value1 = new_value1.split()
	if (len(split_value1) > 1):
		a = split_value1[0]
		b = split_value1[1:]
		b[:] = [x for x in b if x != a]
		if (len(b) > 0):
			new_value1 = a + " " + ' '.join(b)
		else:
			new_value1 = a
	names_dict[original_value] = new_value1
	names_set.add(new_value1)
```

```python
print "len(dict):\t" + str(len(names_dict)) + "\nlen(set):\t" + str(len(names_set)) + "\ndiff of:\t" + str(len(names_dict) - len(names_set))
```

```python
# MAKE THE FIRST WORD OF A MULTIWORD ITEM THE VALUE OF THE KEY MULTIWORD ITEM, IF THE FIRST WORD IS ACTUALLY A LONE ENTRY IN VALUE1 LIST
value1_list = list(names_set)
value1_list = sorted(list(value1_list))
value1_dict = {}
for value1 in value1_list:
	value1_split = value1.split()
	if (len(value1_split) > 0):
		if (value1_split[0] in value1_list) & (value1_split[0] != 'VITAMIN'):
			value1_dict[value1] = value1_split[0]
```

```python
value2_set = set(value1_dict.values())
value2_list = sorted(list(value2_set))


value_list_df = pd.DataFrame({'L1':value2_list[1:], 'L2':value2_list[0:-1]})
def LRatio(x):
	return Levenshtein.ratio(x[0],x[1])

value_list_df['LRatio'] = pd.Series(value_list_df.apply(LRatio, axis=1))
```
